### **1. Create Dummy Data**

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some dummy data
# We'll create 50 points that follow roughly y = 2x + 10
X = 2 * np.random.rand(50, 1) # 50 random numbers for X between 0 and 2
y = 10 + 2 * X + np.random.randn(50, 1) # y = 10 + 2x + some random noise

# Let's plot our data to see what it looks like
plt.scatter(X, y)
plt.xlabel("X (Input Feature)")
plt.ylabel("y (Target Value)")
plt.title("Our Dummy Data")
plt.grid(True)
plt.show()
```

### **2. Implement the Linear Regression Algorithm**

```python
# --- Hyperparameters ---
# These are settings we choose for our algorithm
learning_rate = 0.1  # How big of a step we take
epochs = 1000        # How many times we go through the data

# --- Parameters ---
# These are the values our algorithm will learn
# Start with random guesses for m and b
m = np.random.randn(1,1)
b = np.random.randn(1,1)

# Get the number of data points
n = float(len(X))

# --- Training Loop (Gradient Descent) --- ðŸ“‰
cost_history = [] # To store the cost at each epoch for plotting

for i in range(epochs):
    # 1. Make a prediction with the current m and b
    y_predicted = m * X + b

    # 2. Calculate the cost (Mean Squared Error)
    # This tells us how wrong we are.
    cost = (1/n) * np.sum((y_predicted - y)**2)
    cost_history.append(cost)
 
    # 3. Calculate the gradients (the direction to step)
    # This uses calculus, but these are the formulas for MSE:
    dm = (2/n) * np.sum(X * (y_predicted - y))
    db = (2/n) * np.sum(y_predicted - y)

    # 4. Update the parameters (take a step downhill)
    m = m - learning_rate * dm
    b = b - learning_rate * db

    # Optional: Print the cost every 100 epochs to see progress
    if (i+1) % 100 == 0:
        print(f"Epoch {i+1}: Cost = {cost:.4f}")

# --- Results ---
print("\n--- Training Complete ---")
print(f"The learned value for m is: {m.item():.4f}")
print(f"The learned value for b is: {b.item():.4f}")
# The original values were m=2 and b=10. Our results should be close!
```

### **3. Visualize the Result**

```python
# Plot the original data
plt.scatter(X, y, label='Original Data')

# Plot the regression line we found
plt.plot(X, m * X + b, color='red', linewidth=3, label='Our Regression Line')

plt.xlabel("X (Input Feature)")
plt.ylabel("y (Target Value)")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)
plt.show()
```

You can also plot the cost_history to see the error decreasing over time, confirming that our algorithm was learning successfully!

```python
# Plotting the cost history
plt.plot(range(epochs), cost_history)
plt.xlabel("Epochs")
plt.ylabel("Cost (MSE)")
plt.title("Cost Reduction Over Time")
plt.grid(True)
plt.show()
```
